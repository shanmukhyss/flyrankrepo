## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

The unit of analysis is one content page. Each row in the dataframe represents a single pseudonymized content item together with its observable search, content, and engagement metrics collected over the trailing 90-day period.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.


import pandas as pd

url = "https://raw.githubusercontent.com/shanmukhyss/flyrankrepo/main/data/raw/content_refresh_anonymized.csv"

df = pd.read_csv(url)

df["is_declining_label"] = (
    df["trend_direction"]
      .str.lower()
      .eq("down")
      .astype(int)
)

print("Shape:", df.shape)
print("Rows represent:", "One content page")

display(df.head())
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A simple rule such as "old content with many impressions" considers only one or two signals at a time. In reality, content performance depends on multiple interacting factors such as content age, update frequency, search impressions, click-through rate, average search position, engagement, and content length. These relationships are difficult to express as a single if-statement and may differ across pages and clients. Signal analysis and machine learning help identify these complex patterns objectively while remaining interpretable.
